**Handling Imbalanced Dataset**
A Dataset with an unequal class Distribution


```python
# Importing the Dependencies
import numpy as np
import pandas as pd
```

```python
# Loading the Dataset into Pandas Dataframe
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
# First 5 rows of the DataFrame
credit_card_data.head()
```

```python
# Distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is highly Imbalanced Datset

0 --> Transactions are Legit
1 --> Transactions are Fraudulent

```python
# Seprating the Fraudlent and Legit Transaction
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Applying the Undersampling to handle Imbalanced Data

**Building a sample dataset containing the similar distribution of legit and Fraudlent Transaction**

Number of Fraudlent Transaction -> 492

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

**Concatenate the two DataFrames**

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```
